```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

### We will read the data and display them, using the data from part 2.

```python
df = pd.read_csv('prepared_data.csv')
df.index.names = ['ID'] # put name ID in index 
df
```

##  Data cleansing
we'll identify anything that can be remarked on the data 

We're going to remove the collones that aren't currently helping us and store them elsewhere. 

```python

# Check for missing values
print(df.isnull().sum())
# drop collumns 
dfclean = df.copy()
# Reorganizing columns
new_order = ['year' , 'month','startClusterName' , 'startClusterZip', 'startClusterID', 'endClusterName','endClusterZip', 'endClusterID',  'weekday', 'daytime','isHoliday',  'distance',  'count',  'total']
dfclean = dfclean[new_order]

bin_labels_month = ['First semester','second Semester']
dfclean['month'] = pd.cut(x = dfclean['month'],
                              bins=[1,6,12],
                              labels = bin_labels_month)
replacer_dict = {'First semester' : 0, 
                 'second Semester' : 1}
dfclean['month'] = dfclean['month'].replace(to_replace = replacer_dict)
dfclean['month'].unique()

dfclean
```

### Discretization
We have a dataset made up of numerical and non-numerical values. We'll start by touching certain numerical values and assigning them certain characteristics:

hour in(morning, noon, evening)

holiday in( True, False)

month in ( First semester, second Semester)


```python
#Let's make a copy of our database to transform this
dfBins = dfclean.copy()

bin_labels_hour = ['night','morning', 'noon','evening']
dfBins['daytime'] = pd.cut(x = dfclean['daytime'],
                              bins=[0,6, 12,18,24],
                              labels = bin_labels_hour)

#dfBins['holiday'] = dfclean['holiday'].astype(bool) if we want to change types of holidays 

dfBins['isHoliday'] = dfclean['isHoliday'].map({1: 'True', 0: 'False'})


dfBins
```

## Numerical coding and norming
Let's try to work on the normalized data while preserving our dataset Let's make a copy . We'll transform the day columns into ( 1,2,3...7)

then, using MinMaxScaller, we'll normalize them to [0,1].

```python
#  Data transformation: Normalize numerical columns using Min-Max scaling
scaler = MinMaxScaler()
dfNumeric = dfclean.copy()
# Transform weekday in numerical 
dfNumeric['weekday'] = dfNumeric['weekday'].replace (to_replace=['Mo','Di','Mi','Do','Fr','Sa','So'], value =[1,2,3,4,5,6,7])

numerical_cols = dfNumeric.select_dtypes(include=np.number).columns
dfNumeric[numerical_cols] = scaler.fit_transform(dfNumeric[numerical_cols])

display(dfNumeric.head())

#Let's check that everything has been done correctly and that the values are between 0 and 1.

dfNumeric.describe().loc[['min','max']]
```

## Binear Coding 
We're going to transform our data into binary form so that it can be easily explained and understood by future machine learning models that can't directly process textual or categorical data.

```python
df_encoded = pd.get_dummies(dfclean, drop_first=True)
df_encoded =df_encoded.replace({True : 1,False: 0})
df_encoded.head()
```

```python

df_encoded = pd.DataFrame(scaler.fit_transform(df_encoded), columns = df_encoded.columns, index = df_encoded.index)
df_encoded
```

```python
df_encoded.to_csv('learning_data.csv', index=False)
```
